
### NYC Public School SAT Performance Analysis

Every year, American high school students take SATs, standardized tests measuring literacy, numeracy, and writing skills. With each section having a maximum score of 800 points, SATs are crucial for students and colleges during the admissions process.

Analyzing school performance is vital for various stakeholders, including policymakers, educators, researchers, and parents deciding on their children's education.

In this analysis, we delve into the SAT performance of New York City (NYC) public schools using the provided dataset, `schools.csv`. Our goal is to address three key questions regarding NYC public school SAT performance.


```python
# URL of the CSV file
csv_url = "https://raw.githubusercontent.com/hasperjiga/Data_Analyst_Portfolio/main/schools.csv"

# Import pandas library
import pandas as pd

# Read in the data from the CSV file using the provided URL
schools = pd.read_csv(csv_url)

# Preview the first few rows of the DataFrame
schools.head()

```

```python
# Question: Which NYC schools have the best math results?

# Calculate the minimum score required for best math results (80% of the maximum score)
best_math = 800 * 0.8

# Filter schools with average math scores equal to or greater than the best math score
math_list = schools[schools['average_math'] >= best_math]

# Select only the 'school_name' and 'average_math' columns from the filtered schools
best_math_schools = math_list[['school_name', 'average_math']]

# Sort the schools based on average math scores in descending order
best_math_schools = best_math_schools.sort_values('average_math', ascending=False)

# Display the DataFrame containing the best performing schools for math
best_math_schools
```

```python
# Question: What are the top 10 performing schools based on the combined SAT scores?

# Calculate the total SAT score for each school by summing the average scores for math, reading, and writing
schools['total_SAT'] = schools['average_math'] + schools['average_reading'] + schools['average_writing']

# Select the 'school_name' and 'total_SAT' columns from the dataset and sort by total SAT scores in descending order
top_10_schools = schools[['school_name', 'total_SAT']].sort_values('total_SAT', ascending=False).head(10)

# Display the DataFrame containing the top 10 performing schools based on combined SAT scores
top_10_schools
```

```python
# Question: Which single borough has the largest standard deviation in the combined SAT score?

# Calculate the number of schools in each borough
num_schools = schools['borough'].value_counts()

# Calculate the average SAT score for each borough
average_SAT = round(schools.groupby('borough')['total_SAT'].mean(), 2)

# Calculate the standard deviation of SAT scores for each borough
std_SAT = round(schools.groupby('borough')['total_SAT'].agg('std'), 2)

# Create a dictionary to store the number of schools, average SAT scores, and standard deviations for each borough
std_dev = {
    'num_schools': num_schools,
    'average_SAT': average_SAT,
    'std_SAT': std_SAT
}

# Create a DataFrame to store the results
largest_std_dev = pd.DataFrame(std_dev)

# Filter the DataFrame to include only the borough with the largest standard deviation
largest_std_dev = largest_std_dev[largest_std_dev['std_SAT'] == largest_std_dev['std_SAT'].max()].reset_index()

# Rename the index column to 'borough'
largest_std_dev = largest_std_dev.rename(columns={'index': 'borough'})

# Display the DataFrame containing the borough with the largest standard deviation in SAT scores
largest_std_dev
```
