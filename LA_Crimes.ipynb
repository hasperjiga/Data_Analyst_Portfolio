```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv('/content/crimes.csv', parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

```python
# Question: Which hour has the highest frequency of crimes?

# Extract hour from the 'TIME OCC' column and convert it to integer
crimes['hour'] = crimes['TIME OCC'].str[:2].astype(int)

# Count the frequency of crimes for each hour
crime_hour = crimes['hour'].value_counts()

# Find the hour with the highest frequency of crimes
peak_crime_hour = crime_hour.idxmax()

# Print the hour with the highest frequency of crimes
print("The hour with the highest frequency of crimes is:", peak_crime_hour)
```

```python
## Question: Which area has the largest frequency of night crimes (crimes committed between 10pm and 3:59am)?

## Change TIME OCC to integer and create filter
# Convert 'TIME OCC' column to integer for comparison
crimes['TIME OCC'] = crimes['TIME OCC'].astype(int)

# Define lower and upper time boundaries for night crimes (10pm to 3:59am)
lower_time = (1200 + 1000)  # 10pm in military time
upper_time = 359  # 3:59am in military time

## Filter crimes by night time
# Filter crimes occurring during night time
night_time_crime = crimes.loc[(crimes['TIME OCC'] > lower_time) | (crimes['TIME OCC'] < upper_time)]

# Count the frequency of crimes for each area during night time
night_crime_location = night_time_crime.groupby('AREA NAME')['AREA NAME'].agg('count')

# Find the area with the largest frequency of night crimes
peak_night_crime_location = night_crime_location.idxmax()

# Print the area with the largest frequency of night crimes
print("The area with the largest frequency of night crimes is:", peak_night_crime_location)
```

```python
## Question: Identify the number of crimes committed against victims of different age groups

## Create labels and bins
# Define age group labels
labels = ['0-17', '18-25', '26-34', '35-44', '45-54', '55-64', '65+']

# Find the maximum age in the dataset
max_age = crimes['Vict Age'].max()

# Define bins for different age groups
bins = [0, 17, 25, 34, 44, 54, 64, max_age]

# Create a new column 'age_class' by categorizing 'Vict Age' into age groups
crimes['age_class'] = pd.cut(crimes['Vict Age'], labels=labels, bins=bins)

# Count the frequency of crimes for each age group
victim_ages = crimes['age_class'].value_counts()

# Print the number of crimes committed against victims of different age groups
print("Number of crimes committed against victims of different age groups:")
print(victim_ages)
```
