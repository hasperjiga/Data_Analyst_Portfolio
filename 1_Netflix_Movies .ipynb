```python
# Raw URL of the dataset
raw_url = https://raw.githubusercontent.com/hasperjiga/Data_Analyst_Portfolio/main/netflix_data.csv

# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt


# Read the Netflix dataset into a DataFrame
netflix_df = pd.read_csv('netflix_data.csv', index_col=0)

# Filter the data to remove TV shows
netflix_subset = netflix_df[netflix_df['type'] != 'TV Show']

# Select columns of interest for movie analysis
netflix_movies = netflix_subset.loc[:, ['title', 'country', 'genre', 'release_year', 'duration']]
```

```python
# Filter movies shorter than 60 minutes
short_movies = netflix_movies[netflix_movies['duration'] < 60]
```

```python
# Assign colors based on genre for visualization
colors = []
for x, row in netflix_movies.iterrows():
    if row['genre'] == 'Children':
        color = 'blue'
    elif row['genre'] == 'Documentaries':
        color = 'green'
    elif row['genre'] == 'Stand-Up':
        color = 'red'
    else:
        color = 'yellow'
    colors.append(color)
```

```python
# Create a scatter plot of movie duration by release year
fig = plt.figure(figsize=(12, 8))
plt.scatter(netflix_movies['release_year'], netflix_movies['duration'], c=colors)
plt.xlabel('Release year')
plt.ylabel('Duration (min)')
plt.title('Movie Duration by Year of Release')
plt.show()
```
